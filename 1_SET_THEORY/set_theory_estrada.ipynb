```python
from matplotlib_venn import venn3
import matplotlib.pyplot as plt

# Given data
total_cars = 25
A = 15
R = 12
W = 11
A_W = 5
A_R = 9
R_W = 4
A_R_W = 3

# Calculate the number of cars with each option only
W_only = W - R_W - A_W + A_R_W
A_only = A - A_W - A_R + A_R_W
R_only = R - A_R - R_W + A_R_W

# Calculate the number of cars with both R and W but not A
R_W_not_A = R_W - A_R_W

# Calculate the number of cars with both A and R but not W
A_R_not_W = A_R - A_R_W

# Calculate the number of cars with only one of the options
one_option = W_only + A_only + R_only

# Calculate the number of cars with at least one option
at_least_one = total_cars - (W_only + A_only + R_only)

# Calculate the number of cars with none of the options
none_option = total_cars - at_least_one

# Create the Venn diagram
venn = venn3(subsets=(A_only, R_only, A_R_not_W, W_only, A_W, R_W, A_R_W),
             set_labels=('A', 'R', 'W'))

# Display the Venn diagram
plt.title("Venn Diagram of Car Options")
plt.show()

# Print the results
print("Number of cars with W only:", W_only)
print("Number of cars with A only:", A_only)
print("Number of cars with R only:", R_only)
print("Number of cars with both R and W but not A:", R_W_not_A)
print("Number of cars with both A and R but not W:", A_R_not_W)
print("Number of cars with only one option:", one_option)
print("Number of cars with at least one option:", at_least_one)
print("Number of cars with none of the options:", none_option)
```
